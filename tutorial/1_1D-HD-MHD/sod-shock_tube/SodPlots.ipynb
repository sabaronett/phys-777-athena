# 1D Shock Tube Plots

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import sys
sys.path.insert(0, '/home/stanley/github/PrincetonUniversity/athena-public-version/vis/python')
import athena_read
```

## Import & Inspect Keys

```python
data = athena_read.tab('Sod.block0.out1.00025.tab')
data.keys()
```

## Density & Pressure

```python
fig, axl = plt.subplots(figsize=(6,5))

axl.set_title('Density & Pressure at $t={:.3f}$'.format(data['time']), size='xx-large')
axl.set_xlabel(r'$x$ / pc', size='large')
axl.set_ylabel(r'$\rho$ / $10^{-5}$ g·cm$^{-3}$', size='large')
axl.xaxis.set_minor_locator(mticker.AutoMinorLocator())
axl.yaxis.set_minor_locator(mticker.AutoMinorLocator())
axl.set_xlim([-0.5, 0.5])
axl.set_ylim([0, 1.1])
axl.grid()
axl.plot(data['x1v'], data['rho'], label=r'$\rho$')
axr = axl.twinx()
axr.set_ylabel(r'$P$ / $10^{-4}$ Ba', size='large')
axr.yaxis.set_minor_locator(mticker.AutoMinorLocator())
axr.set_ylim([0, 1.1])
axl.plot(data['x1v'], data['press'], label=r'$P$')
axl.legend(loc='upper right')
plt.show()
```
